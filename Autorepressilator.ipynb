```python
%pylab inline
import numpy as np
import scipy as sp
from scipy import integrate
```

### Один авторепресселятор

```python
def f1(y, t):
    Kalpha = 1
    Kgamma = 6
    Kdegr = 0.2
    ydot = np.zeros_like(y)
    
    ydot[0] = Kalpha/(1 + y[2]**Kgamma) - Kdegr*y[0]
    ydot[1] = Kalpha/(1 + y[0]**Kgamma) - Kdegr*y[1]
    ydot[2] = Kalpha/(1 + y[1]**Kgamma) - Kdegr*y[2]
    
    return ydot
```

```python
t = np.linspace(0, 100, 1000)
y0 = np.array([0, 0, 0.1])
y = integrate.odeint(f1, y0, t)

figure(figsize=(16, 8))
subplot(121)
plot(t, y)
subplot(122)
plot(y[:,0], y[:,1])
show()
```

### Два авторепресселятора

```python
def f2(y, t):
    Kalpha = 1
    Kgamma = 6
    Kdegr = 0.3
    Ktransp = 0.01
    ydot = np.zeros_like(y)
    
    ydot[0] = Kalpha/(1 + y[2]**Kgamma) - Kdegr*y[0] + Ktransp*y[3] - Ktransp*y[0]
    ydot[1] = Kalpha/(1 + y[0]**Kgamma) - Kdegr*y[1] + Ktransp*y[4] - Ktransp*y[1]
    ydot[2] = Kalpha/(1 + y[1]**Kgamma) - Kdegr*y[2] + Ktransp*y[5] - Ktransp*y[2]
    
    ydot[3] = Kalpha/(1 + y[5]**Kgamma) - Kdegr*y[3] + Ktransp*y[0] - Ktransp*y[3]
    ydot[4] = Kalpha/(1 + y[3]**Kgamma) - Kdegr*y[4] + Ktransp*y[1] - Ktransp*y[4]
    ydot[5] = Kalpha/(1 + y[4]**Kgamma) - Kdegr*y[5] + Ktransp*y[2] - Ktransp*y[5]
    
    return ydot
```

```python
t = np.linspace(0, 100, 1000)
y0 = np.array([0, 0, 0.1, 
               0, 0, 0.2])
y = integrate.odeint(f2, y0, t)

figure(figsize=(18, 6))
subplot(131)
plot(t, y[:,:3])
subplot(132)
plot(t, y[:,3:])
subplot(133)
plot(y[:,0], y[:,1], 'r', y[:,3], y[:,4], 'b')
show()
```
